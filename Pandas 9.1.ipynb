```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
#nan mean not number value
```

```python
df = pd.DataFrame(people)

df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
df
```

```python
#remove the those data 
df.dropna()
```

```python
#dropna() using some defoult arguments in the background
#if set 'index' drops rows, set 'columns' then drops columns
#how argument tell to drop if any data missing in that row, drop the row
df.dropna(axis='index', how='any')
```

```python
df.dropna(axis='index', how='all')
```

```python
df.dropna(axis='columns', how='all')
```

```python
df.dropna(axis='index', how='any', subset=['email'])
#jut will drop if their email column is empty
```

```python
df.dropna(axis='index', how='all', subset=['last', 'email'])
#as long as last name and email there it shount drop the rows
```

```python
df.isna()
```

```python
df.fillna(0)
#it is quite usefull with num data
```
