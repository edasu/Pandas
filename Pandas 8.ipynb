```python
country_grp['LanguageWorkedWith'].str.contains('Python').sum()
#read the erorr, because its retun objects it is not series so it says use the apply method 
#using the apply methode we ned do def function (like lambda) that run on every in this group
```

```python
country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
#after this piont we need the % of the number
country_repodents = df['Country'].value_counts()
country_kwn_pythn = country_grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_repodents
```

```python
#we can combine more than on series together using pandas concat method
python_df = pd.concat([country_repodents, country_kwn_pythn], axis='columns', sort=False)
python_df 
```

```python
python_df.rename(columns={'Country': 'NumRepondents', 'LanguageWorkedWith': 'NumKnowsPython'},inplace=True)
```

```python
python_df
```

```python
#dont forget adding column like that cool
python_df['PctKnowPython'] = (python_df['NumKnowsPython'] / python_df['NumRepondents']) * 100
python_df
```

```python
#sort this result , saot the countries larges pct who knows python
python_df.sort_values(by='PctKnowPython', ascending=False, inplace=True)
```
